```python
from langchain.text_splitter import MarkdownHeaderTextSplitter
from langchain.document_loaders import Docx2txtLoader
```

```python
loader = Docx2txtLoader('Introduction_to_Data_and_Data_Science mark.docx')
document = loader.load()
```

```python
print(document[0])
```

```python
page_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=[('#','Title'), ('##','Lecture')])
pages = page_splitter.split_text(document[0].page_content)
pages
```

```python
for page in pages:
    page.page_content = ' '.join(page.page_content.split())
```
